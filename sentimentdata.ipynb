```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sentiment_data.csv")
```

```python
df_cleaned = df.drop(columns=['Unnamed: 0'])
```

```python
df_cleaned = df_cleaned.dropna(subset=['Comment'])
```

```python
sentiment_counts = df_cleaned['Sentiment'].value_counts().sort_index()
```

```python
label_map = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}
sentiment_counts_named = sentiment_counts.rename(index=label_map)
```

```python
colors = ['crimson', 'gold', 'seagreen']

plt.figure(figsize=(6,6))
plt.pie(
    sentiment_counts_named,
    labels=sentiment_counts_named.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=(0.05, 0.05, 0.05)
)
plt.title('Sentiment Distribution (Pie Chart)')
plt.tight_layout()
plt.show()
```
